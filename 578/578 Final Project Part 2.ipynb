```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn import tree
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

sns.set_theme(style="ticks",color_codes=True)
```

```python
dataset_filename = "./data/adult.data.csv"
df_all = pd.read_csv(dataset_filename, delimiter=',', header=None,
                 names=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
                        'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
                        'hours-per-week', 'native-country','income'], skipinitialspace=True)

print(df_all)
```

```python
columns = ['age', 'education', 'education-num', 'income']
df = df_all[columns]
# print(df)
# print('-=' * 30 + '-')

dfLow = df[df['income'] == '<=50K'][['age', 'education', 'education-num']]
# print(dfLow)
# print('-=' * 30 + '-')

dfHigh = df[df['income'] == '>50K'][['age', 'education', 'education-num']]
# print(dfHigh)
# print('-=' * 30 + '-')

fig, ax = plt.subplots(figsize = (12, 6))

agesHigh = np.array(dfHigh['age'])
eduNumsHigh = np.array(dfHigh['education-num'])
scale = 250.0 * np.random.rand(len(agesHigh))
ax.scatter(agesHigh, eduNumsHigh, s = scale, c = 'yellow', alpha=0.5, edgecolors='none', label='>50K')

agesLow = np.array(dfLow['age'])
eduNumsLow = np.array(dfLow['education-num'])
scale = 250.0 * np.random.rand(len(agesLow))
ax.scatter(agesLow, eduNumsLow, s = scale, c = 'blue', alpha=0.5, edgecolors='none', label='<=50K')

ax.set_xlabel('Age')
ax.set_ylabel('Education')
ax.legend()

plt.show()
```

```python
columns = ['age', 'capital-gain', 'income']
df = df_all[columns]
# print(df)
# print('-=' * 30 + '-')

dfLow = df[df['income'] == '<=50K'][['age', 'capital-gain']]
# print(dfLow)
# print('-=' * 30 + '-')

dfHigh = df[df['income'] == '>50K'][['age', 'capital-gain']]
# print(dfHigh)
# print('-=' * 30 + '-')

fig, ax = plt.subplots(figsize = (6, 10))

agesHigh = np.array(dfHigh['age'])
capHigh = np.array(dfHigh['capital-gain'])
scale = 250.0 * np.random.rand(len(agesHigh))
ax.scatter(agesHigh, capHigh, s = scale, c = 'yellow', alpha=0.5, edgecolors='none', label='>50K')

agesLow = np.array(dfLow['age'])
capsLow = np.array(dfLow['capital-gain'])
scale = 250.0 * np.random.rand(len(agesLow))
ax.scatter(agesLow, capsLow, s = scale, c = 'blue', alpha=0.5, edgecolors='none', label='<=50K')

ax.set_xlabel('Age')
ax.set_ylabel('Capital-gain')
ax.legend()

plt.show()
```
