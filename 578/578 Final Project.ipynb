```python
# Import and 

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn import tree
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

sns.set_theme(style="ticks",color_codes=True)
```

```python
dataset_filename = "./data/adult.data.csv"
df_all = pd.read_csv(dataset_filename, delimiter=',', header=None,
                 names=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
                        'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
                        'hours-per-week', 'native-country','income'], skipinitialspace=True)

print(df_all)
```

```python
columns_sex_income = ['sex', 'income']

df = df_all.groupby(columns_sex_income).size().reset_index().rename(columns={0:'count'})
# print(df)
# print('-=' * 30 + '-')

dfHigh = df[df['income']=='>50K'].sort_values(by=['sex'])
dfLow = df[df['income']=='<=50K'].sort_values(by=['sex'])

columnNames = np.array(dfHigh['sex'])

fig, ax = plt.subplots()
bottom = np.zeros(2)

p1 = ax.bar(columnNames, np.array(dfLow['count']), 0.4, label='<=50K', bottom=bottom)
bottom += np.array(dfLow['count'])

p2 = ax.bar(columnNames, np.array(dfHigh['count']), 0.4, label='>50K', bottom=bottom)

ax.legend(loc='upper left')
ax.set_ylabel('Number of people of different income')

plt.title('Number of <=50K or >50K income by gender')
plt.show()
```

```python
columns_education = ['education', 'education-num', 'income']

df = df_all.groupby(columns_education).size().reset_index().rename(columns={0: 'count'}).sort_values(by=['education-num'])
# print(df)
# print('-=' * 30 + '-')

dfLow = df[df['income'] == '<=50K'][['education', 'count']]
# print(dfLow)
# print('-=' * 30 + '-')

dfHigh = df[df['income'] == '>50K'][['education', 'count']]
# print(dfHigh)
# print('-=' * 30 + '-')

edus = np.array(dfLow['education'])

labelLoc = np.arange(len(edus))
width = 0.45
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

offset = width * multiplier
rects = ax.barh(labelLoc + offset, np.array(dfLow['count']), width, label='<=50K')
# ax.bar_label(rects, padding=3)
multiplier += 1

offset = width * multiplier
arr = np.array(dfHigh['count'])
arr = np.insert(arr, 0, 0, axis=0)
rects = ax.barh(labelLoc + offset, arr, width, label='>50K')
# ax.bar_label(rects, padding=3)
multiplier += 1

ax.set_yticks(labelLoc + width, np.array(dfLow['education']))
ax.legend(loc='upper right', ncols=3)
ax.set_xlabel('Number of people with different income')

plt.title('Number of <=50K or >50K income by education')
plt.show()
```

```python
columns_occupation = ['occupation', 'income']

df = df_all.groupby(columns_occupation).size().reset_index().rename(columns={0: 'count'}).sort_values(by=['occupation'])
# Filter out ? occupation
df = df[df['occupation'] != '?']
# print(df)
# print('-=' * 30 + '-')

dfLow = df[df['income'] == '<=50K'][['occupation', 'count']]
# print(dfLow)
# print('-=' * 30 + '-')

dfHigh = df[df['income'] == '>50K'][['occupation', 'count']]
# print(dfHigh)
# print('-=' * 30 + '-')

occupations = np.array(dfLow['occupation'])

labelLoc = np.arange(len(occupations))
width = 0.4
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

offset = width * multiplier
rects = ax.bar(labelLoc + offset, np.array(dfLow['count']), width, label='<=50K')
# ax.bar_label(rects, padding=3)
multiplier += 1

offset = width * multiplier
rects = ax.bar(labelLoc + offset, np.array(dfHigh['count']), width, label='>50K')
# ax.bar_label(rects, padding=3)
multiplier += 1

ax.set_xticks(labelLoc + width, np.array(dfLow['occupation']))
ax.legend(loc='upper right', ncols=3)
ax.set_ylabel('Number of people with different income')

plt.title('Number of <=50K or >50K income by occupation')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
columns_age = ['age', 'income']
df = df_all.groupby(columns_age).size().reset_index().rename(columns={0: 'count'}).sort_values(by=['age'])
# print(df)
# print('-=' * 30 + '-')

dfLow = df[df['income'] == '<=50K'][['age', 'count']]
# print(dfLow)
# print('-=' * 30 + '-')

dfHigh = df[df['income'] == '>50K'][['age', 'count']]
# print(dfHigh)
# print('-=' * 30 + '-')

minAge = min(np.concatenate((np.array(dfLow['age']), np.array(dfHigh['age']))))
maxAge = max(np.concatenate((np.array(dfLow['age']), np.array(dfHigh['age']))))

countsLow = [0]*(maxAge - minAge + 1)
countsHigh = [0]*(maxAge - minAge + 1)

for _, row in dfLow[['age', 'count']].iterrows():
    countsLow[row['age']-17] = row['count']

for _, row in dfHigh[['age', 'count']].iterrows():
    countsHigh[row['age']-17] = row['count']

plt.plot(np.arange(minAge, maxAge+1), countsLow, label='<=50K')
plt.plot(np.arange(minAge, maxAge+1), countsHigh, label='>50K')

plt.title('Relationship between age and income')
plt.xlabel('Age')
plt.ylabel('Number of people with different income')
plt.legend()
plt.show()
```

```python
# Pie chart of education
columns_education = ['education']
df = df_all.groupby(columns_education).size().reset_index().rename(columns={0: 'count'}).sort_values(by=['count'])
# print(df)
# print('-=' * 30 + '-')


plt.pie(df['count'], labels=df['education'])

totalNum = sum(df['count'])
print(totalNum)
labels = ['{0} - {1:1.2f} %'.format(row['education'], 100 * row['count'] / float(totalNum))  for _, row in df.iterrows()]
plt.legend(bbox_to_anchor=(1.3, 0.5), labels=labels, loc='center left', alignment='right')
plt.subplots_adjust(left=0.0, bottom=0.1, right=0.45)
plt.show()
```
